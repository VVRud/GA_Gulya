## Тип репродукції

Задається параметрами `keep_parents` і `keep_elitism` 

`keep_elitism` - кількість найкращих особин, що гарантовано переходять в наступне покоління.

0 <= `keep_elitism` <= `sol_per_pop`

Значення `keep_parents` ігнорується, якщо заданий `keep_elitism` > 0.

Якщо `keep_parents` = -1, тоді всі відібрані в батьківський пул особини переходять в наступне покоління.

Інакше, 0 <= `keep_parents` <= `num_parents_mating`

Місця в наступному поколінні, що залишились, будуть заповнені нащадками.


## Умови зупинки:

- Досягнення максимальної кількості поколінь (параметр `num_generations`).
- За допомогою `stop_criteria`:
  - `reach_40` - алгоритм зупиниться, коли функція пристосування досягне 40. Для багатоцільової оптимізації: `reach_40_20`
  - `saturate_7` - алгоритм зупиниться, коли впродовж 7ми поколінь не змінювалось пристосування.
  
  Може приймати список критеріїв: `stop_criteria=['reach_10_20_30', 'reach_90_-5.7_10']`

Користувач може реалізувати власні умови зупинки за допомогою параметру `on_generation`. Передана функція викликається в кінці кожного покоління, і якщо повертає значення `stop` - алгоритм зупиняється.

````python
def func_generation(ga_instance):
    if ga_instance.best_solution()[1] >= 70:
        return "stop"


# Алгоритм зупиниться, якщо значення функції пристосування перевищить 70.
ga_instance = pygad.GA(on_generation=func_generation)
````

## Функція пристосованості
Задається функціональним параметром `fitness_func`. 

Передана функція повинна приймати на вхід три параметри: 
- обʼєкт класу `pygad.GA` (на якому викликається `run()`)
- хромосома (або список хромосом, якщо `fitness_batch_size` > 1)
- індекс хромосоми в популяції (або список індексів, якщо `fitness_batch_size` > 1)

````python
# Функція пристосування, обернена до суми квадратів генів.
def fitness_func(ga_instance, solution, solution_idx):
    fitness = 1.0 / np.sum(solution**2)
    return fitness
````

# Оператори відбору

Метод відбору особин в батьківський пул задається параметром `parent_selection_type`. 
За замовчуванням обирається `sss` - Steady-state selection - обирання n найкращих батьків.
Інші доступні опції: 
- `rws` - roulette wheel selection - метод рулетки
- `sus` - stochastic universal selection
- `rank` - рангова селекція
- `random` - випадкова селекція
- `tournament` - турнірний відбір

````python
# Задання розміру батьківського пулу і методу відбору.
ga_instance = pygad.GA(parent_selection_type="sss", num_parents_mating=4)

# Для турнірного відбору є можливість вказати кількість учасників турніру.
# За замовчуванням вона дорівнює 3.
ga_instance = pygad.GA(
    parent_selection_type="tournament", num_parents_mating=4, K_tournament=3
)
````

### Можливість реалізації власних операторів.

Окрім вказаних методів вибору батьків, в бібліотеці PyGAD є можливість запрограмувати власні оператори відбору і передати їх у якості параметра до конструктора `pygad.GA`.

Вимоги до функції: приймає на вхід три параметри:

- Список значень функції пристосувань для поточної популяції.
- Кількість батьків, що потрібно відібрати.
- Обʼєкт класу `pygad.GA` - його можна використовувати для отримання доступу до інших атрибутів ГА, таких як популяція.

Повертати функція повинна два обʼєкти:
- відібрані батьки - numpy array розмірністю [кількість батьків x кількість генів]
- індекси відібраних батьків в популяції - 1D numpy array

````python
# Steady-state selection. Сортування і відбір найбільш пристосованих особин.
def parent_selection_func(fitness, num_parents, ga_instance):
    fitness_sorted = sorted(range(len(fitness)), key=lambda k: fitness[k])
    fitness_sorted.reverse()

    parents = np.empty((num_parents, ga_instance.population.shape[1]))
    for parent_num in range(num_parents):
        parents[parent_num, :] = ga_instance.population[
            fitness_sorted[parent_num], :
        ].copy()

    return parents, np.array(fitness_sorted[:num_parents])


ga_instance = pygad.GA(
    parent_selection_type=parent_selection_func, num_parents_mating=4
)
````

# Оператори кросинговеру

Оператор кросинговеру задається параметром `crossover_type`. 
За замовчуванням обирається `single_point` - Одноточковий кросинговер.
Інші доступні опції: 
- `two_points` - двоточковий кросинговер
- `uniform` - при цьому параметрі для кожного гена випадковим чином обирається один з двох відповідних генів батьків.
- `scattered` - логіка така ж як у попередньому типі, але має більш оптимізовану реалізацію, використовуючи векторизовані операції NumPy
- `None` - етап кросинговеру пропускається

````python
# Задання ймовірності і методу кросинговеру.
ga_instance = pygad.GA(crossover_type="scattered", crossover_probability=0.7)

# Відключення кросинговеру.
ga_instance = pygad.GA(crossover_type=None)
# або
ga_instance = pygad.GA(crossover_probability=0.0)
````

### Можливість реалізації власних операторів.
Аналогічно до методів селекції, в бібліотеці PyGAD є можливість визначення власних операторів кросинговеру.

Вимоги до функції: приймає на вхід три параметри:

- Обрані батьки
- Очікувана розмірність результату - tuple (`offspring_size`, `num_genes`)
- Обʼєкт класу `pygad.GA` - його можна використовувати для отримання доступу до інших атрибутів ГА, таких як популяція.

Функція повинна повернути
- offspring - numpy array розмірністю, що передана другим параметром.

````python
# Одноточний кросинговер.
def crossover_func(parents, offspring_size, ga_instance):
    offspring = []
    idx = 0
    while len(offspring) != offspring_size[0]:
        parent1 = parents[idx % parents.shape[0], :].copy()
        parent2 = parents[(idx + 1) % parents.shape[0], :].copy()

        random_split_point = np.random.choice(range(offspring_size[1]))

        parent1[random_split_point:] = parent2[random_split_point:]

        offspring.append(parent1)

        idx += 1

    return np.array(offspring)


ga_instance = pygad.GA(crossover_type=crossover_func, crossover_probability=0.5)
````

# Оператори мутації

Метод мутації задається параметром `mutation_type`.
За замовчуванням обирається `random` - додавання випадкового значення до гену.
`mutation_by_replacement` - булевий параметр, що використовується тільки з `random` мутацією.
Якщо True - ген повністю замінюється випадковим значення, інакше - випадкове значення додається до гену.
Інший параметри, що працюють тільки з випадковою мутацією: `random_mutation_min_val` і `random_mutation_max_val` - задають рівномірний розподіл.
Інші доступні опції: 
- `swap` - зміна генів місцями
- `inversion` - обирання підмножини генів і збереження їх в оберненому порядку
- `scramble` - обирання підмножини генів і збереження їх у випадковому порядку
- `adaptive` - адаптивна мутація при якій у особин з нижчою пристосованістю ймовірність мутації вища.
- `None` - етап мутації пропускається

````python
# Задання ймовірності мутації
ga_instance = pygad.GA(mutation_probability=0.1)
# або
ga_instance = pygad.GA(mutation_percent_genes=10)
# або
ga_instance = pygad.GA(mutation_num_genes=1)  # Мутація рівно одного гена


# Випадкова заміна генів на число від -0.5 до 0.5.
ga_instance = pygad.GA(
    mutation_type="random",
    mutation_probability=0.1,
    mutation_by_replacement=True,
    random_mutation_min_val=-0.5,
    random_mutation_max_val=0.5,
)

# Задання двох ймовірностей при адаптивній мутації.
ga_instance = pygad.GA(mutation_type="adaptive", mutation_probability=[0.1, 0.2])
````

### Можливість реалізації власних операторів.
Аналогічно до попередніх операторів, існує можливість визначення власних операторів мутації.

Вимоги до функції: приймає на вхід два параметри:

- Створені нащадки (після кросинговеру)
- Обʼєкт класу `pygad.GA` - для доступу до параметрів і атрибутів алгоритму.

Функція повинна повернути
- offspring - змінені нащадки - такої ж розмірності як перший параметр.

````python
# Мутація рівно одного гена в кожному нащадку
def mutation_func(offspring, ga_instance):
    for chromosome_idx in range(offspring.shape[0]):
        random_gene_idx = np.random.choice(range(offspring.shape[1]))

        offspring[chromosome_idx, random_gene_idx] += np.random.random()

    return offspring


ga_instance = pygad.GA(mutation_type=mutation_func)
````

# Оператори інверсії

Окремо в фреймворку етап інверсії не виділяють. Розробники пропонують виконувати інверсії на етапі мутації.
Є можливість обрати тип мутації `inversion`. Або ж перевизначити свій тип мутації, що буде виконувати інверсію генів разом з додатковими мутаціями.

# Інші вбудовані можливості

## Багатоцільова оптимізація

Підтримується багатоцільова оптимізація. Для цього функція передана в якості параметра `fitness_func` повинна повертати масив значень.

В роботі з багатоцільовою оптимізацію допомагає модуль `pygad.utils.nsga2.NSGA2`. 

## Multimodal optimization

Не знайшов в документації, але в коді бачив функції, до дозволяють обчислювати відстань до сусідів (crowding_distance), тощо. Для того щоб урізноманітнити популяцію і знаходити одразу декілька локальних максимумів варто перевизначати оператори селекції або функцію пристосування: надавати перевагу особам, що знаходяться далеко від інших.

## Динамічна оптимізація

Динамічна оптимізація можлива за допомогою PyGAD, оскільки всі функції, що може визначити користувач приймають на вхід також обʼєкт самого алгоритму (`pygad.GA`). Це означає, що функції можуть змінювати логіку в залежності від, наприклад, номеру покоління.

Також можна змінювати певні параметри алгоритму у процесі роботи, як наприклад розмір популяції. В документації є окремий розділ, що описує як уникати помилок в роботі алгоритму при ручній зміні параметрів.

## Вирішення недетермінованих задач

Фреймворк підходить для вирішення недетермінованих задач. Але автори попереджають, що в такому випадку необхідний набір параметрів:

```
keep_elitism=0
keep_parents=0
keep_solutions=False
keep_best_solutions=False
```
Таким чином фреймворк не буде перевикористовувати вже обчислені значення функції пристосування.

## Сумісність з моделями keras i torch

Підмодулі `pygad.kerasga` i `pygad.torchga` дають змогу використовувати генетичний алгоритм для тренування моделей, створених за допомогою вказаних фреймворків. 

Модулі `pygad.nn`, `pygad.cnn`, `pygad.gann` - дозволяють створювати штучні нейронні мережі і тренувати їх за допомогою ГА.

**Самі автори зазначають, що фреймворк не оптимізований для тренування великих моделей і буде використовувати забагато ресурсів**

## Дебаг

Фреймворк передбачає, що користвувач може передати свої функції-колбеки, що будуть викликатись алгоритмом на відповідних кроках.

Цей механізм можна використовувати як для зміни поведінки алгоритму, так і для розширеного аналізу його роботи.

````python
def on_start(ga_instance):
    print("GA started!")
    print(f"===Population: \n {np.round(ga_instance.population, 2)}")


def on_fitness(ga_instance, calc_fitness):
    print(f"=Generation {ga_instance.generations_completed + 1} started!")
    print(f"==Fitness: {calc_fitness}")
    return calc_fitness


def on_parents(ga_instance, parents):
    print("==Parents selected!")
    print(f"===Selected parents: \n {np.round(parents, 2)}")


def on_crossover(ga_instance, offspring):
    print("==Crossover finished!")
    print(f"===After crossover: \n {np.round(offspring, 2)}")
    return offspring


def on_mutation(ga_instance, offspring):
    print("==Mutation finished!")
    print(f"===After mutation: \n {np.round(offspring, 2)}")
    return offspring


def on_stop(ga_instance, fitness):
    print(f"GA stopped at generation {ga_instance.generations_completed}!")
    print(f"==Best fitness: {fitness}")


def on_generation(ga_instance):
    print(f"=Generation : {ga_instance.generations_completed} finished!\n")
    rounded_population = np.round(ga_instance.population, 2)
    print(f"Population : \n {rounded_population} )")
````

## Тестові задачі
Розробником надані приклади для оцінки ГА:
- лінійна оптимізація
- багатоцільова лінійна оптимізація
- відтворення картинки

# Демонстрація роботи:
Хромосома складається з генів - дійсних чисел.

Задача: мінімізувати довжину вектора, представленого генами.

````python
def fitness_func(ga_instance, solution, solution_idx):
    fitness = 1.0 / np.sum(solution**2)
    return fitness
````

Інші параметри:

````python
num_generations = 20
num_genes = 2
fitness_function = fitness_func

# Ініціалізація випадкових значень для генів
init_range_low = -2
init_range_high = 2

# В популяції 3 особини, з них 2 стають батьками і виживають.
sol_per_pop = 4
num_parents_mating = 2
keep_parents = 2

# Вільне місце в популяції (1) буде займати наступне покоління

# Метод рулетки для відбору батьків
parent_selection_type = "rws"

# Метод одноточкового кросовера
crossover_type = "single_point"
crossover_probability = 0.7

# Метод мутації - додавання випадкового числа до гена
mutation_type = "random"
mutation_percent_genes = 50
````

Ініціалізація ГА:

````python
ga_instance = pygad.GA(
    num_generations=num_generations,
    num_parents_mating=num_parents_mating,
    fitness_func=fitness_func,
    sol_per_pop=sol_per_pop,
    num_genes=num_genes,
    init_range_low=init_range_low,
    init_range_high=init_range_high,
    parent_selection_type=parent_selection_type,
    keep_parents=keep_parents,
    keep_elitism=0,
    on_start=on_start,
    on_fitness=on_fitness,
    on_parents=on_parents,
    on_crossover=on_crossover,
    on_mutation=on_mutation,
    on_generation=on_generation,
    on_stop=on_stop,
    crossover_type=crossover_type,
    crossover_probability=crossover_probability,
    mutation_type=mutation_type,
    mutation_percent_genes=mutation_percent_genes,
)
````

````python
ga_instance.run()
````

````python
ga_instance.plot_fitness()
````

# Висновки

## Переваги
- хороша документація

Детально описані параметри, різноманітні їх комбінації. Добре описані додаткові функції. До всього дуже багато коментарів в коді самої бібліотеки.
- висока можливість кастомізації

Можливо задавати власні оператори селекції, кросинговеру і мутації. Також є можливість надавати callback функції що будуть викликатись на специфічних етапах роботи алгоритму.
- простота у використанні

Все що потрібно для використання - створити обʼєкт `pygad.GA` з потрібними параметрами і запустити алгоритм.


## Недоліки
- погана оптимізація для великих задач

Неозброєним оком видно, що код генетичних операторів не оптимальний, часто відбуваються зайві проходи по популяції. Видно що самі автори в курсі проблеми, але не зрозуміло, чому замість оптимізації створюють модулі для інтеграції з фреймворками машинного навчання.
- архітектура бібліотеки

Практично вся логіка знаходиться в одному класі - `pygad.GA`. Це робить читання і розуміння коду дуже складною задачею, до того ж неможливо використати якісь окремі частини фреймворку - обернена сторона простоти використання.
- дуже багато параметрів конструктора

Вони часто неявно перекривають інші параметри, деякі комбінації параметрів приводять до неочікуваних результатів. Автори приділяють цьому багато уваги в документації, але на практиці дуже легко пропустити певні нюанси.
